## 4.3 データの前処理

````python
#@title **コード 4-13** 特徴量を変数 xに代入
xcol = ['身長', '体重', '年代']
x = df[xcol]
x
````

````python
#@title **コード 4-14** 正解データを参照してtに代入
t = df['派閥']
t
````

## 4.4 モデルの準備と学習の実行

````python
#@title **コード 4-15** tree モジュールのインポート
from sklearn import tree
````

````python
#@title **コード 4-16** モデルの準備と学習の実行
# モデル（決定木モデル）の準備（未学習）
model = tree.DecisionTreeClassifier(random_state = 0)

# 学習の実行
model.fit(x, t)
````

````python
#@title **コード 4-17** きのこかたけのこか予測する
taro = [[170, 70, 20]]

model.predict(taro)
````

※ 2次元リストで特徴量データを渡しているので警告がでている

````python
#@title **コード 4-18** 複数の予測を一度に実行

matsuda = [172, 65, 20]
asagi = [158, 48, 20]

new_data = [matsuda, asagi]

model.predict(new_data)
````

#### <font color= "#a00"><b>学習済みモデルでの予測</b></font>
```Python
モデル変数.predict(特徴量のデータ)
```

## 4.5 モデルの評価

````python
#@title **コード 4-19**
model.score(x, t)
````

#### <font color= "#a00"><b>モデルの正解率を計算</b></font>
```Python
モデル変数.score(特徴量のデータ, 正解のデータ)
```

## 4.6 モデルの保存

````python
#@title **コード 4-20** モデルの保存

import pickle

with open('KinokoTakenoko.pkl', 'wb') as f:
  pickle.dump(model, f)
````

````python
#@title **コード 4-21**  KinokoTakenoko.pkl からモデルを変数に読み込む

import pickle

with open('KinokoTakenoko.pkl', 'rb') as f:
  model2 = pickle.load(f)
````
